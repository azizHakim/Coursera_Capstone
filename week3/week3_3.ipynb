### 3. Create new dataframe consisting of Borough, Neighbourhood, Latitude and Longitude

```python
neighborhoods=df.drop(['Postcode'],axis=1)
neighborhoods
```

### 4. Print number of boroughs and neighbourhoods

```python
print('The dataframe has {} boroughs and {} neighborhoods.'.format(
        len(neighborhoods['Borough'].unique()),
        neighborhoods.shape[0]
    )
)
```

### 5. Create a map of Toronto with neighborhoods superimposed on top

```python
map_toronto = folium.Map(location=[43.65, -79.4], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(neighborhoods['Latitude'], neighborhoods['Longitude'], neighborhoods['Borough'], neighborhoods['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
